```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/Titanic/train.csv')
```

```python
df.head()
```

# 1. Categorical Data

## a. Countplot

```python
sns.countplot(df, x = 'Pclass')
```

```python
df['Survived'].value_counts().plot(kind = 'bar')
```

## b. Pie Chart

```python
df['Sex'].value_counts().plot(kind = 'pie', autopct = '%.1f')
```

# Numerical Data

## a. Histogram

```python
# Create bins in data range

plt.hist(df['Age'], bins = 10)
```

## b. Distplot

```python
# KDE - Kernel Density Estimation
# PDF - Probability Density Functions

sns.distplot(df['Age'])
```

## c. Boxplot

```python
# Gives 5 number summary(to recognize outliers and noise)
# Min, 25 percentile, median , 75 percentile, max
#      (Interquartile                  Range)

sns.boxplot(df, x = 'Age')
```

```python
print(df['Age'].min())
print(df['Age'].max())
print(df['Age'].mean())
```

# Skewness

```python
# if 0 perfectly balanced
# if +ve skewed to right
# if -ve skewed to left

df['Age'].skew()
```
