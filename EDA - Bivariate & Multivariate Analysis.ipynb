# 1. Scatter plot (Numerical - Numerical)

```python
# Bivariate
sns.scatterplot(tips, x = 'total_bill', y = 'tip')
```

```python
# Multi variate
sns.scatterplot(tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', size = 'size')
```

# 2. Bar plot(Numerical - Categorical)

```python
# x-axis --> Categorical data
# y-axis --> Numerical data

sns.barplot(titanic, x = 'Pclass', y = 'Age', hue = 'Sex')

# Black bar shows confidence interval
```

# 3. Box Plot(Numerical - Categorical)

```python
sns.boxplot(titanic, x = 'Sex', y = 'Age', hue = 'Survived')
```

# 4. Distplot(Numerical - Categorical)

```python
sns.displot(titanic, x = 'Age', hue = 'Survived', kde = True, stat = 'density', linewidth = 0)
```

```python
sns.histplot(titanic[titanic['Survived']==0]['Age'],kde=True, stat="density", linewidth=0,fill=False)
sns.histplot(titanic[titanic['Survived']==1]['Age'],kde=True, stat="density", linewidth=0,fill=False)
```

# 5. Heat Map(Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
print(titanic['Survived'].dtype)
```

```python
(titanic.groupby('Embarked')['Survived'].mean())*100
```

# 6. Cluster Map(Categorical - Categorical)

```python
pd.crosstab(titanic['Parch'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```
