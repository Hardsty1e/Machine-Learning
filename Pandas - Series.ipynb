```python
#how to create a Series oject

llist = ['ind', 'pak', 'usa', 'eng', 'aus']

cou = pd.Series(llist)
```

```python
#custom index

llist = ['ind', 'pak', 'usa', 'eng', 'aus']
runs = [67, 57, 89, 100, 99]

pd.Series(runs, index = llist)
```

```python
#setting a name for series object to call in future

marks = pd.Series(runs, index = llist, name = 'Countries of World')
marks
```

```python
#using dictionary to create series

marks = {
    'maths' : 67,
    'english' : 57,
    'science' : 89,
    'hindi' : 100
}

marks_series = pd.Series(marks, name = 'Subject marks')
marks_series
```
